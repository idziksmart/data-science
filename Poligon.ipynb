# Standaryzacja - Standard Scaler
doprowadzic do takiej postaci gdzie skala nie ma znaczenia ale znaczenie ma wariancja (rozrzut)

odchylenie standardowe to estymator obciazony wariancji

std() - pandas nieobciazaony
std() - numpy obciazony

```python
# Odchylenie standardowe określa, o ile – średnio biorąc – jednostki zbiorowości różnią się od średniej arytmetycznej badanej zmiennej.
# Im zbiorowość jest bardziej zróżnicowana, tym wariancja, a więc i odchylenie standardowe, są większe.
# wariancja to odchylenie standardowe podniesione do kwadratu
```

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
# proces standaryzacji to odjecie od kazdej wartosci sredniej wartosci i podzielenie przez odchylenie standardowe
```

```python
def standardize(x):
    return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
# standaryzacja w sklearn, roznica ze wzgledu na obciazenie
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
# ten sposob jest polecany
# kolejny sposob plus zachowanie tych wystandaryzowanych wartosci
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
dane = scaler.transform(df[['price']])
```

```python
# standaryzacja prowadzi do ustawienie sredniej = 0
dane.mean()
```

```python
# standaryzacja prowadzi do ustawienie odchylenia standarowego = 1
dane.std()
```

```python
# w ten sposob kilka kolumn na raz
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```

## Uzupelnienie brakow SimpleImputer

```python
from sklearn.impute import SimpleImputer

# strategy: 'mean', 'median', 'most_frequent', 'constant' - sposob w jaki braki danych zostana uzupelnione srednia, mediana, najczesciej powtarzajace sie, stala wartosc
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df[['weight']])
```

```python
imputer.statistics_
```

```python
df[['weight']]
```

```python
imputer.transform(df[['weight']])
```

```python
df['weight'] = imputer.transform(df[['weight']])
df
```

```python
# constant - stala wartosc
imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=99.0)
imputer.fit_transform(df[['price']])
```

Wypełnienie braków metodą forward fill
